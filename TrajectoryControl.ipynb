Real-time control of steer wheel angle for semi-autonomous ground vehicles
==========================================================================

Abstract
========
Neste trabalho, se visa o controle em tempo real da posição lateral do veículo. Para o controle, atua-se sobre o ângulo do volante, então, esta variável será nossa sinal de controle que irá modificar a trajetória do veículo para atingir a posição de referência. Na Figura 1 mostra-se o diagrama de controle.

```python
from IPython.display import SVG
SVG(filename='./Trabalho/Control.svg')
```

* Na Figura, a posição lateral (Desired Path) é nossa referência, o erro medido serve como entrada para o controlador (Steer Angle Controller). A sinal de controle (Steer Angle) computada por o controlador é enviada ao atuador (volante) para modificar a trajetória do veículo, obtendo-se assim a nova posição lateral do veículo. Este laço de simulação é feito até o tempo de simulação determinado por o usuario, para nosso caso é 30 segundos.

1. PD Controller
================

Para o controle se disenho um controlador PD, da forma:

```python
import IPython.display as dsp
```

```python
dsp.Math(r"u(t) = kp(x(t)_{ref} - x(t)) + kd(\dot{x}(t)_{ref} - \dot{x}(t))")
```

Na qual, a posição de referência, velocidade de referência e a variação da posição lateral são:

```python
dsp.Math(r"x(t)_{ref} = -1.6, \text{ }\dot{x}(t)_{ref} = 0 \text{ e } \dot{x}(t) = \frac{x(t) - x(t-1)}{dt}")
```

2. Vehicle Dynamics
===================

Para a simulação da dinâmica do veículo, utilizou-se o [CarSim](http://carsim.com/), o qual envia por TCP/IP a posição lateral atual do veículo ao controlador [Raspberry PI](http://www.raspberrypi.org/) que computa o control PD e envia a sinal de controle (Steer Wheel Angle) de volta para o simulador. O processo é repetido cada delta de tempo (0.001 s).

3. Results
==========

```python
import pandas as pd
from pylab import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

filename = './Trabalho/file.txt'
data = pd.read_csv(filename)
```

Lateral Position [m]
====================

```python
fig, axes = plt.subplots(figsize=(12,6), dpi=100)
# Plot data of control
axes.plot(data.time, data.latPos, 'b', label='Current Lateral Position', lw=2)

# Plot reference position
line, = axes.plot(linspace(0,30,100), -1.6 * ones((100,1)), 'k', label='Reference', lw=1.5)
line.set_dashes([5, 10, 15, 10])

axes.legend(loc=2)
axes.set_xlabel('Time [s]', fontsize=15)
axes.set_ylabel('Lateral Position [m]', fontsize=15)
grid()
show()
```

Variation of lateral position [m/s]
===================================

```python
fig, axes = plt.subplots(figsize=(12,6), dpi=100)
# Plot data of control
axes.plot(data.time, data.velPos, 'b', lw=2)

axes.set_xlabel('Time [s]', fontsize=15)
axes.set_ylabel('Variation of Lateral Position [m/s]', fontsize=15)
grid()
show()
```

Signal control (Steer wheel angle) [deg]
========================================

```python
fig, axes = plt.subplots(figsize=(12,6), dpi=100)
# Plot data of control
axes.plot(data.time, data.signalC, 'b', lw=2)

axes.set_xlabel('Time [s]', fontsize=15)
axes.set_ylabel('Signal control (Steer Wheel Angle) [deg]', fontsize=15)
grid()
show()
```

Global position of vehicle [m]
==============================

```python
fig, axes = plt.subplots(figsize=(12,6), dpi=100)
# Plot data of control
axes.plot(data.xPos, data.yPos, 'b', lw=2)

axes.set_xlabel('X Position [m]', fontsize=15)
axes.set_ylabel('Y Position [m]', fontsize=15)
grid()
show()
```
